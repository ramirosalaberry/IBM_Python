Bajo la pagina web de wikipedia (https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M), la guardo en la carpeta de labs y luego la abro con Beautiful soup.
Download the wikipedia website (https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M), I save it in the labs folder and then open it with Beautiful soup.


```python
with open('List of postal codes of Canada_ M - Wikipedia.html') as fp:
    soup = BeautifulSoup(fp, "html.parser")
```

Rescato la tabla de la pagina HTML guardada y la convierto en Data Frame.
I rescue the table from the saved HTML page and convert it to Data Frame.

```python
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
df.tail(10)
```

```python
df.shape
```

Proceso el data frame tabla quitando cada uno de las celdas y armando el data frame final
Process the table data frame by removing each of the cells and assembling the final data frame

```python
columns = ['PostalCode', 'Borough', 'Neighborhood']
df_toronto = pd.DataFrame(columns=columns)
pos_cod=[]
bor=[]
nei=[]
for f in range(len(df)):
    for c in range(len(df.columns)):
        valor=str(df.loc[f,c])
        postal_code=valor[0:3]
        resto=valor[3:]
        splitresto= re.split('[()/]',resto)
        munic=splitresto[0]
        splitresto[1:]
        vecin=[]
        for i in splitresto[1:]:
            a=(i.lstrip()).rstrip()
            if  a:
                vecin.append(a)
            else:
                pass
        pos_cod.append(postal_code)
        bor.append(munic)
        valor_nei=', '.join(vecin)
        nei.append(valor_nei)
        
df_toronto['PostalCode']=pos_cod
df_toronto['Borough']=bor
df_toronto['Neighborhood']=nei
#Busco los municipio no asignados a un postal code
df_toronto[df_toronto['Borough']=='Not assigned'].head(20)
```

Elimino las filas que tienen un municipio no asignado y reseteo el indice
I delete the rows that have an unassigned Borough and reset the index

```python
indexNames = df_toronto[ (df_toronto['Borough'] == 'Not assigned') ].index
df_toronto.drop(indexNames , inplace=True)
df_toronto = df_toronto.reset_index(drop=True)
df_toronto.sort_values(by='PostalCode')
df_toronto
```
